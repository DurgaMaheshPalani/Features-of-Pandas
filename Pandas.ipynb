# Slicing & Indexing

## Explicit Indexes

```python
# read the weather data and print it
import pandas as pd

## Start the code here
weather=pd.read_csv(r"E:\DATA SCIENCE\data sets\weather.csv")
weather
## End here
```

```python
#1 Show the name of columns.
## Start the code here
weather.columns
## End here
```

```python
#2 Set RainToday as index value
## Start the code here
weather_index=weather.set_index("RainToday")
weather_index
## End here
```

```python
#3 Reindex the index value to its previous state.
## Start the code here
weather.reindex()
## End here
```

```python
#1  Subset Rainfall value from (0.0 to 3.6)
## Start the code here
Rainfall = [0.0, 3.6]
weather.Rainfall.isin(Rainfall)
## End here
```

```python
#2 Set "RainTomorrow" as an index value
## Start the code here
weather_index1=weather.set_index("RainTomorrow")
weather_index1
## End here
```

```python
#3 subset its 'Yes' value.
## Start the code here
weather[weather.RainTomorrow =='Yes']
## End here
```

```python
#4 Set RainToday & RainTomorrow as multilevel index.
## Start the code here
weather_index2=weather.set_index(["RainToday","RainTomorrow"])
weather_index2
## End here
```

```python
#5 Subset outer level keeping its value 'No'.
## Start the code here
weather_index2.loc[('No',slice(None)),:]
## End here
```

```python
#6 Subset inner levels with list of tuples keeping the values ("No","Yes") ("Yes","No")
## Start the code here
weather_index2.loc[(slice(None),['No','Yes']),:]
## End here
```

```python
#7 Sort the table by index value
## Start the code here
weather_index2.sort_index()
## End here
```

## Slicing & Subsetting with .loc & .iloc 

```python
#1 Set the weather index as "MinTemp" and "MaxTemp" then sort it.
## Start the code here
weather.set_index(['MinTemp','MaxTemp']).sort_index()
## End here
```

```python
#Slice outer index level

#Q2. Slice the MinTemp from (6.1 to 8.0)
## Start the code here
weather.set_index(['MinTemp','MaxTemp']).query('MinTemp>=6.1 and MinTemp<=8.0')
## End here
```

```python
#Slice the inner index level

#Q3. Slice the first four index i.e, (6.1,18.2):(8.0,24.3)
## Start the code here
weather.set_index(['MinTemp','MaxTemp']).loc[([6.1,8.0],[18.2,24.3]),:]
## End here
```

```python
#Slicing columns

#Q4. Slice weather by columns (Rainfall and WindGustSpeed)
## Start the code here
weather[['Rainfall','WindGustSpeed']]
## End here
```

```python
#5 Slice weather twice from (6.1, 18.3 to 8.0,24.3) keeping columns RainToday and RainTomorrow
## Start the code here
weather.loc[(weather['MinTemp'].isin([6.1,8.0])) & (weather['MaxTemp'].isin([18.3,24.3])),:]
## End here
```

```python
#Slicing by dates

#Q6. Set "Date" as index , sort() it and Slice date between 2020-01-01 and 2020-04-10
## Start the code here
weather.set_index('Date').sort_index().query('Date>="2020-01-01" and Date<="2020-04-10"')
## End here
```

```python
#7 Subset weather by row (2 to 5) and columns (1 to 4)
## Start the code here
weather.iloc[2:6,1:5]
## End here
```



## Working with pivot tables


```python
#1 Pivot the weather by keeping value as "Rainfall", index as "MinTemp" and columns as "MaxTemp".
## Start the code here
weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall')
## End here
```

```python
#2 Slice first five index
## Start the code here
weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall').iloc[:5]
## End here
```

```python
#3 Calculate summary stats across index and columns separately, i.e (mean)
## Start the code here
weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall',aggfunc='mean')
## End here
```

# Creating & Visualizing Dataframes

## Visualizing your data

```python
# import matplotlib and pandas 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the weather data
weather1=pd.read_csv("E:\DATA SCIENCE\data sets\weather.csv")
weather1.head()
```

```python
#1 Plot histogram of "Rainfall" keeping bins as 25 and show it.
## Start the code here
weather1['Rainfall'].plot.hist(bins=25)
## End here
```

```python
#Bar plots

#Q2. Assign "Rainfall" & "RainToday" to df and perform bar plot.
## Start the code here
df=sns.barplot(x='RainToday',y='Rainfall',data=weather1)
plt.show()
## End here
```

```python
#Line plot

#Q3. Assign "RainToday" and "WindGustSpeed" to df1 and perform line plot
## Start the code here
df1=sns.lineplot(x='RainToday',y="WindGustSpeed",data=weather1)
plt.show()
## End here
```

```python
#Layering plots

#Q4. Layer the plots keeping the dataframe as "RainTomorrow" & "Rainfall" and also add a legend("Yes","No")
## Start the code here
df2=weather1[['RainTomorrow','Rainfall']]
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='Yes', label='Yes')
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='No', label='No')
plt.legend()
plt.show()
## End here
```

```python
#5 Add Transperancy , i.e alpha of 0.5
## Start the code here
df2=weather1[['RainTomorrow','Rainfall']]
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='Yes', label='Yes',alpha=0.5)
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='No', label='No',alpha=0.5)
plt.legend()
plt.show()
## End here
```
